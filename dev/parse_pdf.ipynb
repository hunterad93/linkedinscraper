```python
import PyPDF2
import os

# Get a list of all PDF files in the ./data/ directory
pdf_files = [f for f in os.listdir('./data/') if f.endswith('.pdf')]

# Loop through each PDF file
for pdf_file in pdf_files:
    # Open the PDF file in read-binary mode
    with open(f'./data/{pdf_file}', 'rb') as file:
        # Create a PDF file reader object
        pdf_reader = PyPDF2.PdfReader(file)

        # Get the number of pages in the PDF
        num_pages = len(pdf_reader.pages)

        # Initialize an empty string to hold the text
        pdf_text = ''

        # Loop through each page and extract the text
        for page in range(num_pages):
            pdf_page = pdf_reader.pages[page]
            pdf_text += pdf_page.extract_text()

    # Save the extracted text to a .txt file
    with open(f'./data/{pdf_file[:-4]}.txt', 'w') as file:
        file.write(pdf_text)
```

```python
import pandas as pd
import re

# Define the columns for the DataFrame
columns = ['Top Skills', 'Certifications', 'Summary', 'Experience', 'Education', 'Raw Text']

# Initialize an empty DataFrame
df = pd.DataFrame(columns=columns)

# List of files
files = [f for f in os.listdir('./data/') if f.endswith('.txt')]

for file_name in files:
    with open(f'./data/{file_name}', 'r') as file:
        text = file.read()

        # Extract the required information using regular expressions
        top_skills = re.search(r'(?<=Top Skills\n)(.*?)(?=\nCertifications|\nSummary)', text, re.DOTALL)
        certifications = re.search(r'(?<=\nCertifications\n)(.*?)(?=\nSummary\n)', text, re.DOTALL)
        summary = re.search(r'(?<=\nSummary\n)(.*?)(?=\nExperience\n)', text, re.DOTALL)
        experience = re.search(r'(?<=\nExperience\n)(.*?)(?=\nEducation\n)', text, re.DOTALL)
        education = re.search(r'(?<=\nEducation\n)(.*)', text, re.DOTALL)

        # Add the extracted information to the DataFrame
        df = df.append({
            'Top Skills': '|'.join([skill.strip() for skill in top_skills.group(0).split('\n') if skill.strip()]) if top_skills else None,
            'Certifications': '|'.join([cert.strip() for cert in certifications.group(0).split('\n') if cert.strip()]) if certifications else None,
            'Summary': summary.group(0).replace('\n', '|') if summary else None,
            'Experience': '|'.join([exp.strip() for exp in experience.group(0).split('\n') if exp.strip()]) if experience else None,
            'Education': '|'.join([edu.strip() for edu in education.group(0).split('\n') if edu.strip()]) if education else None,
            'Raw Text': text
        }, ignore_index=True)

# Print the DataFrame
df
```
